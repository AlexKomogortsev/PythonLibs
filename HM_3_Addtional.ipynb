```python
# Task 5

X = pd.DataFrame(data['data'], columns=data['feature_names'])
X
```

```python
# Task 6

X.size
X.isna().sum()
```

```python
# Task 7

target = pd.DataFrame({'target': data['target']})
X = pd.concat([X, target], axis=1)
X
```

```python
# Task 8

from matplotlib import pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,10))

X_corr = X.corr()
X_corr = np.round(X_corr, 2)
X_corr[np.abs(X_corr) < 0.3] = 0

sns.heatmap(X_corr, annot=True, linewidth=5, cmap='coolwarm')

plt.title('Correlation Matrix')
plt.show()
```

```python
# Task 9

high_corr = []
for i, j in enumerate(X_corr.get('target')):
    if j > 0.5 and X_corr.index[i] != 'target':
        high_corr.append(X_corr.index[i])
    else:
        continue
high_corr
```

```python
# Task 10

#X = X.drop('target', axis=1)

for i in high_corr:
    X[f'{i}_2'] = X[i].pow(2)
X.describe()
```
